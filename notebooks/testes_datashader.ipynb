```python
import cartopy.crs as ccrs
import geoviews as gv
import xarray as xr
from holoviews.operation.datashader import rasterize
from bokeh.models import CustomJSHover
import intake

gv.extension("bokeh")

catalog = intake.open_catalog('http://ftp1.cptec.inpe.br/pesquisa/das/carlos.bastarz/GSIMonitor/anls/catalog.yml')

dfs = catalog['2024031806'].to_dask()

data = dfs['VIRTUAL TEMPERATURE'].sel(lev=1000)

# `gv.project` is to fix disappearing rasterize plot when pan zooming
# Opts is to look like hvplot
# Reference: https://github.com/holoviz/geoviews/issues/483
ax = rasterize(gv.project(gv.Dataset(data).to(gv.QuadMesh)))
ax.opts(
    projection=ccrs.PlateCarree(),
    cmap="coolwarm",
    tools=["hover"],
    colorbar=True,
    #symmetric=True,
    width=1024,
    height=768,
)

ax

# Adding Coastline
#ax * gv.feature.coastline
```
