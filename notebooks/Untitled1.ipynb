```python
import hvplot.pandas
import geopandas as gpd
import matplotlib.pyplot as plt

zctas_gdf = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

fig, ax = plt.subplots(figsize=(12, 7))
ax.set_aspect('equal')
ax.axis('off')

facecolor = ['red'] * len(zctas_gdf)
facecolor[2] = 'orange'
zctas_gdf.plot(ax=ax, alpha=0.85, facecolor='none', edgecolor='black')

plt.grid(False)
plt.tight_layout()
plt.show()
plt.close()
```
